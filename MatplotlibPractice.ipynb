```python
import numpy as np 
import pandas as pd 
import matplotlib.pyplot as plt
import matplotlib.ticker as tck
```

```python
fig, ax = plt.subplots()

x = np.arange(0,10,1)

ax.plot(x, x, label = 'linear')
ax.plot(x, x ** 2, label = 'quadratic')
ax.plot(x, x ** 3, label = 'cubic')
ax.plot(x, x ** 4, label = 'quartic')
ax.plot(x, x ** 5, label = 'quintic')



#formatting plot
ax.set_xlabel(r'$x$')                   #format the number for equations
ax.set_ylabel(r'$y$') 
ax.grid()                               #format the number for equations
legend = ax.legend(prop={"size":10}, loc = 'upper left')
plt.ylim(0,60000)
ax.get_yaxis().set_major_formatter(tck.FuncFormatter(lambda x, p: format(int(x), ',')))
plt.title('Polynomial Equations', fontweight ='bold', fontsize = 14)


```
